```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs

# Create a sample dataset for demonstration
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Define the Mean Shift clustering model
bandwidth = 2.0  # Bandwidth controls the size of the kernel used for density estimation
ms = MeanShift(bandwidth=bandwidth)

# Fit the model to the data
ms.fit(X)

# Get the cluster centers
cluster_centers = ms.cluster_centers_

# Get the labels for each data point
labels = ms.labels_

# Number of clusters
n_clusters = len(cluster_centers)

print(f'Number of estimated clusters: {n_clusters}')

# Plot the data points and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', marker='X')
plt.title('Mean Shift Clustering')
plt.show()

```
